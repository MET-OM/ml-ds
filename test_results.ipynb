```python
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
x = xr.open_dataset("example/test_input.nc")
y = xr.open_dataset("example/test_target.nc")
yh = xr.open_dataset("example/test_prediction.nc")
```

```python
fig,ax = plt.subplots(1,3,figsize=(20,5))


plot_var = "v10"
time_idx = 0

x[plot_var][time_idx].plot.imshow(ax=ax[0],norm=mcolors.Normalize(-10,10))
y[plot_var][time_idx].plot.imshow(ax=ax[1],norm=mcolors.Normalize(-10,10))
yh[plot_var][time_idx].plot.imshow(ax=ax[2],norm=mcolors.Normalize(-10,10))

ax[0].set_title("Original ERA5 data")
ax[1].set_title("Re-interpolated ERA5 data")
ax[2].set_title("ML-adjusted data")
```
